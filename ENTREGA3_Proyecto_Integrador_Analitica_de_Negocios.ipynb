**Identificación y Tratamiento de Outliers**


*Conversión a numérico:*


Se utiliza *pd.to_numeric()* en la columna *PASAJEROS* con el parámetro *errors='coerce'*, lo que convierte cualquier valor no numérico en NaN.

Eliminación de filas con NaN:
Después de la conversión, se eliminan las filas con valores NaN en la columna *PASAJEROS* usando *dropna(subset=['PASAJEROS'])*. Esto asegura que solo queden valores numéricos en esa columna.

Cálculo del IQR:
Después de limpiar los datos, se calcula el rango intercuartílico (IQR) para identificar los valores extremos (outliers).

Filtrado y eliminación de outliers:
Se eliminan los valores fuera del rango permitido por el IQR, generando un DataFrame limpio de outliers.

```python
import pandas as pd

# Cargar la base de datos (usando 'oper' como el DataFrame de operación de pasajeros)
id = '1BDBecnlRYz0mQ_0LZAb3DVoBnNqYjgcd'
url = f'https://docs.google.com/uc?export=download&id={id}'
oper = pd.read_csv(url, delimiter=';')

# Convertir la columna PASAJEROS a numérica (forzar conversión, eliminando o llenando con NaN los valores no numéricos)
oper['PASAJEROS'] = pd.to_numeric(oper['PASAJEROS'], errors='coerce')

# Eliminar las filas donde PASAJEROS sea NaN, ya que no son válidas para el análisis
oper = oper.dropna(subset=['PASAJEROS'])

# Paso 1: Calcular los cuartiles y el rango intercuartílico (IQR) para la columna PASAJEROS
Q1 = oper['PASAJEROS'].quantile(0.25)  # Primer cuartil (25%)
Q3 = oper['PASAJEROS'].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico

# Paso 2: Definir los límites para identificar los outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Paso 3: Filtrar los outliers (valores fuera del rango)
outliers = oper[(oper['PASAJEROS'] < limite_inferior) | (oper['PASAJEROS'] > limite_superior)]
print(f"Outliers detectados:\n{outliers}")

# Paso 4: Eliminar los outliers del DataFrame original
oper_sin_outliers = oper[(oper['PASAJEROS'] >= limite_inferior) & (oper['PASAJEROS'] <= limite_superior)]

# Verificar el número de filas antes y después de eliminar los outliers
print(f"Número de filas antes de eliminar outliers: {len(oper)}")
print(f"Número de filas después de eliminar outliers: {len(oper_sin_outliers)}")
oper_sin_outliers
```

## Ahora unimos las bases de datos con la función merge para concatenar el ID de los municipios con el nombre se los mismos

```python
import pandas as pd

# **Base de datos 2: Operación de pasajeros y despachos**
id = '1BDBecnlRYz0mQ_0LZAb3DVoBnNqYjgcd'
url = f'https://docs.google.com/uc?export=download&id={id}'
oper = pd.read_csv(url, delimiter=';')

# Asegurarse de que los códigos de municipio tengan 5 dígitos
oper['MUNICIPIO_ORIGEN_RUTA'] = oper['MUNICIPIO_ORIGEN_RUTA'].astype(str).str.zfill(5)
oper['MUNICIPIO_DESTINO_RUTA'] = oper['MUNICIPIO_DESTINO_RUTA'].astype(str).str.zfill(5)

# **Limpiar la columna FECHA_DESPACHO y organizar el formato de HORA_DESPACHO**
# Convertir las fechas al formato deseado YYYY-MM-DD y convertir a datetime para eliminar la parte de la hora
oper['FECHA_DESPACHO'] = pd.to_datetime(oper['FECHA_DESPACHO'], errors='coerce').dt.date

# Organizar el formato de hora
# Asegurarse de que HORA_DESPACHO esté en formato HH:MM
oper['HORA_DESPACHO'] = oper['HORA_DESPACHO'].apply(lambda x: f'{int(x):02}:00' if pd.notnull(x) else '00:00')

# **Base de datos de DIVIPOLÍTICA de municipios**
div = pd.read_excel('DIVIPOLA_Municipios.xlsx', header=10, usecols="C:D").loc[:1121]
div.rename(columns={'Código .1': 'Código', 'Nombre.1': 'Nombre'}, inplace=True)
div['Nombre'] = div['Nombre'].str.title()
div['Código'] = div['Código'].astype(int).astype(str).str.zfill(5)

# **Unir el nombre del municipio de origen con la base de operaciones**
oper = pd.merge(oper, div[['Código', 'Nombre']], left_on='MUNICIPIO_ORIGEN_RUTA', right_on='Código', how='left')
oper.rename(columns={'Nombre': 'nombre_origen'}, inplace=True)

# **Unir el nombre del municipio de destino con la base de operaciones**
oper = pd.merge(oper, div[['Código', 'Nombre']], left_on='MUNICIPIO_DESTINO_RUTA', right_on='Código', how='left')
oper.rename(columns={'Nombre': 'nombre_destino'}, inplace=True)

# **Eliminar columnas duplicadas**
oper.drop(columns=['Código_x', 'Código_y'], inplace=True)
oper

```
